```python
import fastf1 as ff1
import pandas as pd
from fastf1.core import Laps
from fastf1 import plotting
from fastf1 import utils
from matplotlib import pyplot as plt
from timple.timedelta import strftimedelta
import numpy as np
from matplotlib.pyplot import figure
from matplotlib.collections import LineCollection
from matplotlib import cm


fp = ff1.get_session(2023, 'Monza', 'Q')
#fp_laps = fp.load_laps(with_telemetry=True)
fp.load()
```

```python
import matplotlib.pyplot as plt
import fastf1.plotting

ff1.plotting.setup_mpl()

lec_lap = fp.laps.pick_driver('LEC').pick_fastest()
ver_lap = fp.laps.pick_driver('VER').pick_fastest()
sai_lap = fp.laps.pick_driver('SAI').pick_fastest()

lec_tel = lec_lap.get_car_data().add_distance()
ver_tel = ver_lap.get_car_data().add_distance()
sai_tel = sai_lap.get_car_data().add_distance()


ferr_color = ff1.plotting.team_color('FER')
rbr_color = ff1.plotting.team_color('RBR')
ferr2_color = "khaki"

fig1, ax1 = plt.subplots(1)
fig1.set_figheight(10)
fig1.set_figwidth(15)
ax1.plot(lec_tel['Distance'], lec_tel['Speed'], color = ferr_color, label = 'LEC', linewidth=2)
ax1.plot(ver_tel['Distance'], ver_tel['Speed'], color = rbr_color, label = 'VER', linewidth=2)
ax1.plot(sai_tel['Distance'], sai_tel['Speed'], color = ferr2_color, label = 'SAI', linewidth=2)

ax1.set_xlabel('Distance in m', size=20)
ax1.set_ylabel('Speed in Km/h', size=20)

ax1.legend(prop={'size': 20})
ax1.set_title(f"{lec_lap['Driver']} - {ver_lap['Driver']} - {sai_lap['Driver']}", fontweight = "bold", ha='center')

fig1.suptitle(f"Fastest lap comparison\n"f"{fp.event.year} {fp.event.EventName} - {fp.name}\n", size = 20, weight='bold', ha='center')
#plt.subplots_adjust(top=0.85)
fig1.tight_layout()

plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import fastf1.plotting

ff1.plotting.setup_mpl()

lec_lap = fp.laps.pick_driver('LEC').pick_fastest()
ver_lap = fp.laps.pick_driver('VER').pick_fastest()
sai_lap = fp.laps.pick_driver('SAI').pick_fastest()

lec_tel = lec_lap.get_car_data().add_distance()
ver_tel = ver_lap.get_car_data().add_distance()
sai_tel = sai_lap.get_car_data().add_distance()

lec_tel_roggia = lec_tel.loc[lec_tel['Distance'].between(1800, 2200, inclusive=True)]
ver_tel_roggia = ver_tel.loc[ver_tel['Distance'].between(1800, 2200, inclusive=True)]
sai_tel_roggia = sai_tel.loc[sai_tel['Distance'].between(1800, 2200, inclusive=True)]

lec_tel_lesmo = lec_tel.loc[lec_tel['Distance'].between(2200, 3000, inclusive=True)]
ver_tel_lesmo = ver_tel.loc[ver_tel['Distance'].between(2200, 3000, inclusive=True)]
sai_tel_lesmo = sai_tel.loc[sai_tel['Distance'].between(2200, 3000, inclusive=True)]

lec_tel_t1 = lec_tel.loc[lec_tel['Distance'].between(600, 1000, inclusive=True)]
ver_tel_t1 = ver_tel.loc[ver_tel['Distance'].between(600, 1000, inclusive=True)]
sai_tel_t1 = sai_tel.loc[sai_tel['Distance'].between(600, 1000, inclusive=True)]

lec_tel_parab = lec_tel.loc[lec_tel['Distance'].between(5000, 5800, inclusive=True)]
ver_tel_parab = ver_tel.loc[ver_tel['Distance'].between(5000, 5800, inclusive=True)]
sai_tel_parab = sai_tel.loc[sai_tel['Distance'].between(5000, 5800, inclusive=True)]

ferr_color = ff1.plotting.team_color('FER')
rbr_color = ff1.plotting.team_color('RBR')
ferr2_color = "khaki"

fig1, ax1 = plt.subplots(2, 2)
fig1.set_figheight(20)
fig1.set_figwidth(20)

ax1[0,0].plot(lec_tel_t1['Distance'], lec_tel_t1['Speed'], color = ferr_color, label = 'LEC', linewidth=3)
ax1[0,0].plot(ver_tel_t1['Distance'], ver_tel_t1['Speed'], color = rbr_color, label = 'VER', linewidth=3)
ax1[0,0].plot(sai_tel_t1['Distance'], sai_tel_t1['Speed'], color = ferr2_color, label = 'SAI', linewidth=3)

ax1[0,1].plot(lec_tel_roggia['Distance'], lec_tel_roggia['Speed'], color = ferr_color, label = 'LEC', linewidth=3)
ax1[0,1].plot(ver_tel_roggia['Distance'], ver_tel_roggia['Speed'], color = rbr_color, label = 'VER', linewidth=3)
ax1[0,1].plot(sai_tel_roggia['Distance'], sai_tel_roggia['Speed'], color = ferr2_color, label = 'SAI', linewidth=3)

ax1[1,0].plot(lec_tel_lesmo['Distance'], lec_tel_lesmo['Speed'], color = ferr_color, label = 'LEC', linewidth=3)
ax1[1,0].plot(ver_tel_lesmo['Distance'], ver_tel_lesmo['Speed'], color = rbr_color, label = 'VER', linewidth=3)
ax1[1,0].plot(sai_tel_lesmo['Distance'], sai_tel_lesmo['Speed'], color = ferr2_color, label = 'SAI', linewidth=3)

ax1[1,1].plot(lec_tel_parab['Distance'], lec_tel_parab['Speed'], color = ferr_color, label = 'LEC', linewidth=3)
ax1[1,1].plot(ver_tel_parab['Distance'], ver_tel_parab['Speed'], color = rbr_color, label = 'VER', linewidth=3)
ax1[1,1].plot(sai_tel_parab['Distance'], sai_tel_parab['Speed'], color = ferr2_color, label = 'SAI', linewidth=3)



ax1[0,0].set_xlabel('Distance in m', size=10)
ax1[0,0].set_ylabel('Speed in Km/h', size=10)

ax1[0,1].set_xlabel('Distance in m', size=10)
ax1[0,1].set_ylabel('Speed in Km/h', size=10)

ax1[1,0].set_xlabel('Distance in m', size=10)
ax1[1,0].set_ylabel('Speed in Km/h', size=10)

ax1[1,1].set_xlabel('Distance in m', size=10)
ax1[1,1].set_ylabel('Speed in Km/h', size=10)

ax1[0,0].legend(prop={'size': 10})
ax1[0,1].legend(prop={'size': 10})
ax1[1,0].legend(prop={'size': 10})
ax1[1,1].legend(prop={'size': 10})


ax1[0,0].set_title(f"Prima variante", ha='center', fontsize=15)
ax1[0,1].set_title(f"Roggia", ha='center', fontsize=15)
ax1[1,0].set_title(f"Lesmo 1 - 2", ha='center', fontsize=15)
ax1[1,1].set_title(f"Parabolica", ha='center', fontsize=15)


fig1.suptitle(f"Fastest lap comparison\n"f"{fp.event.year} {fp.event.EventName} - {fp.name}\n{lec_lap['Driver']} - {ver_lap['Driver']} - {sai_lap['Driver']}",
               size = 20, weight='bold', ha='center')
#plt.subplots_adjust(top=0.85)
fig1.tight_layout()

plt.show()
```
